```python
merged_focos["data_pas"] = pd.to_datetime(merged_focos["data_pas"], errors="coerce")
merged_focos["data"] = merged_focos["data_pas"].dt.date
merged_focos["hora"] = merged_focos["data_pas"].dt.time

merged_focos = merged_focos[(merged_focos["bioma"]=="Cerrado") & (merged_focos["estado"]=="DISTRITO FEDERAL")].copy()

merged_focos.head()
```

```python
# === 2) filtrar por coordenadas de brasília
# centro aproximado
lat0, lon0 = -15.7939, -47.8828

# opção A: bounding box simples (~0.5º ~ 55 km)
dlat, dlon = 0.5, 0.5
mask_bb = (
    merged_focos["lat"].between(lat0 - dlat, lat0 + dlat)
    & merged_focos["lon"].between(lon0 - dlon, lon0 + dlon)
)
focos_bb = merged_focos[mask_bb].copy()
```

```python

# === 3) agregar por dia (contagem e target)
focos_dia = (
    focos_bb.groupby("data")
            .size()
            .reset_index(name="n_focos")
            .sort_values("data")
)

# garantir todos os dias do intervalo (para criar os 0s)
idx = pd.date_range(focos_dia["data"].min(), focos_dia["data"].max(), freq="D").date
focos_dia = (
    focos_dia.set_index("data")
             .reindex(idx)
             .fillna({"n_focos": 0})
             .rename_axis("data")
             .reset_index()
)
focos_dia["n_focos"] = focos_dia["n_focos"].astype(int)
focos_dia["target"] = (focos_dia["n_focos"] > 0).astype(int)
```

```python
# === 4) merge com meteorologia diária (já processada)
meteo = pd.read_csv(r"C:\Users\sofia\Documents\Data_Analytics_Ironhack\Projects\google_brain\Climate_data\merged_meteo.csv")  # o que criaste com médias/somas
meteo["Data"] = pd.to_datetime(meteo["Data"], errors="coerce").dt.date

# alinhar tipos e juntar
focos_dia["data"] = pd.to_datetime(focos_dia["data"]).dt.date
df_final = focos_dia.merge(meteo, left_on="data", right_on="Data", how="left").drop(columns=["Data"])

# === 5) features históricas úteis (sem leakage)
df_final = df_final.sort_values("data").reset_index(drop=True)
df_final["focos_ultimos_3d"] = df_final["n_focos"].rolling(3, min_periods=1).sum().shift(1)
df_final["focos_ultimos_7d"] = df_final["n_focos"].rolling(7, min_periods=1).sum().shift(1)

# precipitação/temperatura/umidade móveis (ajusta nomes às tuas colunas no meteodiário)
if "precipitacao_mm" in df_final.columns:
    df_final["chuva_7d"] = df_final["precipitacao_mm"].rolling(7, min_periods=1).sum().shift(1)
if "temp_media" in df_final.columns:
    df_final["temp_7d"] = df_final["temp_media"].rolling(7, min_periods=1).mean().shift(1)
if "umidade_media" in df_final.columns:
    df_final["umidade_7d"] = df_final["humidade_media"].rolling(7, min_periods=1).mean().shift(1)

# === 6) temporais
s = pd.to_datetime(df_final["data"])
df_final["mes"] = s.dt.month
df_final["dia_semana"] = s.dt.weekday

```

```python
# salvar
df_final.to_csv("df_brasilia_focos_meteo.csv", index=False)
print(df_final.head(), df_final.shape)
```
