```python
import pandas as pd
```

```python
climate_data_2024 = pd.read_csv(r"C:\Users\sofia\Documents\Data_Analytics_Ironhack\Projects\google_brain\Climate_data\INMET_CO_DF_A001_BRASILIA_01-01-2024_A_31-12-2024.csv",
                                skiprows=8,
                                sep=";",         
                                encoding="latin1")

climate_data_2024.head()
```

```python
climate_data_2024.info()
```

```python
climate_data_2024['Data'] = pd.to_datetime(climate_data_2024['Data'], format="%Y/%m/%d", errors='coerce')

for col in climate_data_2024.columns[2:]: #starting on the third column
    climate_data_2024[col] = (
        climate_data_2024[col].astype(str)
        .str.replace(",", ".")
        .str.strip()
        .replace("", "0")
        .astype(float)
    )

meteo_daily_2024 = climate_data_2024.groupby('Data').agg({
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'sum',
    'RADIACAO GLOBAL (Kj/m²)': 'mean',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'mean',
    'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)': 'max',
    'TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)': 'min',
    'TEMPERATURA DO PONTO DE ORVALHO (°C)': 'mean',
    'UMIDADE RELATIVA DO AR, HORARIA (%)': 'mean',
    'VENTO, RAJADA MAXIMA (m/s)': 'max',
    'VENTO, VELOCIDADE HORARIA (m/s)': 'mean'
}).reset_index()


meteo_daily_2024.rename(columns={
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'precipitacao_mm',
    'RADIACAO GLOBAL (Kj/m²)': 'radiacao_media',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'temp_media',
    'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)': 'temp_max',
    'TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)': 'temp_min',
    'TEMPERATURA DO PONTO DE ORVALHO (°C)': 'temp_orvalho_media',
    'UMIDADE RELATIVA DO AR, HORARIA (%)': 'humidade_media',
    'VENTO, RAJADA MAXIMA (m/s)': 'vento_rajada_max',
    'VENTO, VELOCIDADE HORARIA (m/s)': 'vento_vel_media'
}, inplace=True)

meteo_daily_2024.head()
```

```python
climate_data_2018 = pd.read_csv(r"C:\Users\sofia\Documents\Data_Analytics_Ironhack\Projects\google_brain\Climate_data\INMET_CO_DF_A001_BRASILIA_01-01-2018_A_31-12-2018.csv",
                                skiprows=8,
                                sep=";",         
                                encoding="latin1",
                                dtype={'DATA (YYYY-MM-DD)': str}
                                )

climate_data_2018.head()
```

```python
climate_data_2018['DATA (YYYY-MM-DD)'] = climate_data_2018['DATA (YYYY-MM-DD)'].str.strip()

climate_data_2018['DATA (YYYY-MM-DD)'] = pd.to_datetime(climate_data_2018['DATA (YYYY-MM-DD)'], format="%Y-%m-%d", errors='coerce')

print(climate_data_2018['DATA (YYYY-MM-DD)'].head(), climate_data_2018['DATA (YYYY-MM-DD)'].isna().sum())
```

```python
import numpy as np

climate_data_2018.rename(columns={"DATA (YYYY-MM-DD)": "Data"}, inplace=True)

for col in climate_data_2018.columns:
    if col != "Data":
        climate_data_2018[col] = (
            climate_data_2018[col].astype(str)
            .str.replace(",", ".", regex=False)
            .replace("-9999", np.nan)
            .replace("", np.nan)
        )
        climate_data_2018[col] = pd.to_numeric(climate_data_2018[col], errors="coerce")


for col in climate_data_2018.columns:
    if col != "Data" and climate_data_2018[col].dtype == object:
        climate_data_2018[col] = (
            climate_data_2018[col].astype(str)
            .str.replace(",", ".")
            .str.strip()
            .replace("-9999", np.nan)
            .replace("", np.nan)
        )
        climate_data_2018[col] = pd.to_numeric(climate_data_2018[col], errors="coerce")

# 5. Agregar por dia
meteo_daily_2018 = climate_data_2018.groupby("Data").agg({
    "PRECIPITAÇÃO TOTAL, HORÁRIO (mm)": "sum",
    "RADIACAO GLOBAL (KJ/m²)": "mean",
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)": "mean",
    "TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)": "max",
    "TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)": "min",
    "TEMPERATURA DO PONTO DE ORVALHO (°C)": "mean",
    "UMIDADE RELATIVA DO AR, HORARIA (%)": "mean",
    "VENTO, RAJADA MAXIMA (m/s)": "max",
    "VENTO, VELOCIDADE HORARIA (m/s)": "mean"
}).reset_index()

# 6. Renomear colunas
meteo_daily_2018.rename(columns={
    "PRECIPITAÇÃO TOTAL, HORÁRIO (mm)": "precipitacao_mm",
    "RADIACAO GLOBAL (KJ/m²)": "radiacao_media",
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)": "temp_media",
    "TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)": "temp_max",
    "TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)": "temp_min",
    "TEMPERATURA DO PONTO DE ORVALHO (°C)": "temp_orvalho_media",
    "UMIDADE RELATIVA DO AR, HORARIA (%)": "humidade_media",
    "VENTO, RAJADA MAXIMA (m/s)": "vento_rajada_max",
    "VENTO, VELOCIDADE HORARIA (m/s)": "vento_vel_media"
}, inplace=True)

meteo_daily_2018.head()
```
